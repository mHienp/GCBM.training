```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/cbm_defaults_v1.2.8340.362.db")
```

```python
# Create cursor object
cur = conn.cursor()

# Execute SQL query to get table names
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")

# Get the result set as a list of tuples
result = cur.fetchall()

print("The database has", len(result), "tables.")
```

```python
# Create empty list to hold table names
table_names = [row[0] for row in result]

for table_name in table_names:
    print(table_name)
```

```python
# Loop through each table and load it into a df
df_list = []
for table_name in table_names:
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    globals()[table_name] = df
    df_list.append(df)
```
