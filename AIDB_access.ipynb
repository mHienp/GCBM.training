```python
for table in tables:
    print(table)
```

```python
for table in tables:
    try:
        globals()[table] = mdb.read_table(dbPath, table, dtype_backend="pyarrow")
    except KeyError as e:
        print(f"Error reading table {table}: {e}")
    except ValueError as e:
        print(f"Error reading table {table}: {e}")
```
